```python
from konlpy.tag import Kkma
from konlpy.utils import pprint
kkma = Kkma()
pprint(kkma.sentences(u'안산공업고등학교 컴퓨터과 3학년입니다. 반갑습니다.'))
pprint(kkma.nouns(u'워드 클라우드를 만들기 위해서 Konlpy 라이브러리를 연습하고 있습니다.'))
pprint(kkma.pos(u'간단한 사용방법을 익히고 API 문서를 참조하여 보다 자세한 내용을 살펴보겠습니다.'))
```

```python
from konlpy.tag import Kkma
kkma = Kkma()
print(kkma.morphs(u'안산시에 위치한 안산공업고등학교는 유일한 공업계 특화성 고등학교입니다.'))
print(kkma.nouns(u'안산공업고등학교는 기계과, 전기과, 전자과, 화공과, 디자인과, 컴퓨터과로 6개과로 학년당 15개 학급으로 이뤄져있습니다.'))
print(kkma.pos(u'행복한 학교, 꿈이 이뤄지는 학교, 학생과 선생님이 함께하는 학교입니다.'))
```

+ morphs() : 구문을 형태소로 구문 분석합니다. 형태소란 의미를 가지는 요소로서는 더 이상 분석할 수 없는 가장 작은 말의 단위를 말합니다.
+ nouns() : 명사 추출기.
+ pos() : 포스 태거. HMM(통계적 태거) 기반이며 태그의 확률을 계산합니다.

```python
from bs4 import BeautifulSoup
import requests
from konlpy.tag import Twitter
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
search_word = "핼로윈"
start_num = 1
title_list = []
#print(req)

def get_title(start_num, end_num) :
  
  while start_num <= end_num :
    url = 'https://search.naver.com/search.naver?where=news&sm=tab_jum&query={}&start={}'.format(search_word,start_num)
    req = requests.get(url)
    
    if req.ok :
      html = req.text
      soup = BeautifulSoup(html, "html.parser")
      #print(soup)  
      titles = soup.select('.news_tit')
      for title in titles :
        title_list.append(title['title'])
    start_num += 1
  #print(title_list)
  #print(len(title_list))
get_title(1, 10)

def make_wordcloud(word_count) :
  twitter = Twitter()
  sentences_tag = []
  for sentence in title_list :
    morph = twitter.pos(sentence)
    sentences_tag.append(morph)
    #print(morph)
    #print("-" * 30)
  #print(sentences_tag)
  #print("\n" * 3)

  noun_adj_list = []

  for sentence in sentences_tag :
    for word, tag in sentence :
      if tag in ['Noun', 'Adjective'] :
        noun_adj_list.append(word)
    
  counts = Counter(noun_adj_list)
  tags = counts.most_common(word_count)
  #print(tags)
  
  wc = WordCloud(font_path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf', background_color='white', width=800, height=600)
  print(dict(tags))
  plt.rc('font', family='NanumBarunGothic')
  cloud = wc.generate_from_frequencies(dict(tags))
  plt.figure(figsize=(10,8))
  plt.axis('off')
  plt.imshow(cloud)
  plt.show()


make_wordcloud(10)
```
