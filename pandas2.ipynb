# 타이타닉의 승객 데이터 분석

## 1. CSV 파일 불러오기

+ PassengerId : 승객ID
+ Survived: 생존 여부 `0` 또는 `1`
+ PClass : 티켓 클래스 - '1', '2', '3'
+ Name : 승객의 이름
+ Sex : 성별
+ Age : 나이
+ SibSp : 형제자매 또는 배우자 수
+ Parch : 부모 또는 자녀 수
+ Ticket : 티켓 번호
+ Fare : 운임
+ Cabin : 객실 번호
+ Embarked : 승선항

```python
import pandas as pd
titanic = pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv')
#titanic = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/titanic.csv',encoding='cp949')
#titanic.columns = ['승객ID', '생존여부', '클래스', '이름', '성별', '나이', '형제자매 또는 배우자의 수', '부모 또는 자녀수', '티켓', '운임', '객실', '승선항']
titanic
```

+ DataFrame 열의 이름 변경

```python
titanic.columns = ['승객ID', '생존여부', '클래스', '이름', '성별', '나이', '형제자매 또는 배우자의 수', '부모 또는 자녀수', '티켓', '운임', '객실', '승선항']
titanic
```
