```python

import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ansan.csv', encoding='cp949', index_col=0, thousands = ',')

data.head()
```

```python
data = data.div(data['2022년10월_계_총인구수'] , axis=0)
del data['2022년10월_계_총인구수'], data['2022년10월_계_연령구간인구수'] 

data
```

```python
name = input("원하는 지역의 이름을 입력해주세요 : ")
a = data.index.str.contains(name) #해당 행을 찾아서 해당 지역의 인구 구조를 저장

data2 = data[a]
data2
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family='NanumGothic')

data2.T.plot()
plt.show()
```
